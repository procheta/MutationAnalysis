```python
#Compute Cooccurrence Based on the Comet Input files in a directory

import os

#The code will generate circos plot input files for drawing co-occurrence links. For each file in the directory voc_only a separate
#file will be generated.
directory="voc_only/"
for filename in os.listdir(directory):
        lineage=filename.split(".txt")[0]
        lineageFile="voc_only/"+filename
        cooc_dict={}
        maxVal=0
        maxSub=""
        count=0
        print("Considering the lineage: ",lineage)
        #The input file is the same as the COMET algorithm Input File
        with open(lineageFile,"r") as f:
            for line in f:
                st=line.split("\n")[0].split("\t")
                x=len(st)
                for i in range(1,x-2):
                    for j in range(i+1,x-1):
                        if st[j]=="":
                            continue;
                        key1=st[i]+"#"+st[j]
                        key2=st[j]+"#"+st[i]
                        if key1 in cooc_dict.keys():
                            cooc_dict[key1]=cooc_dict[key1]+1
                            if maxVal < cooc_dict[key1]:
                                maxVal=cooc_dict[key1]
                        elif key2 in cooc_dict.keys():
                            cooc_dict[key2]=cooc_dict[key2]+1
                            if maxVal < cooc_dict[key2]:
                                maxVal=cooc_dict[key2]
                        else:
                            cooc_dict[key1]=1
                            if maxVal < cooc_dict[key1]:
                                maxVal=cooc_dict[key1]
                count=count+1
                
                


        cooc_count=0
        if count < maxVal:
            print(count," ",maxVal," ",lineage)
        #Creating a text file to store the Co-occurrence information
        writeFile="Cooc_Graph/"+lineage+".txt"
        with open(writeFile,"w") as f1:
            for node in cooc_dict.keys():
                cooc_dict[node]=cooc_dict[node]/count
                #Considering Co-occurrence only greater than 80%
                if cooc_dict[node] >= 0.8:
                    nodes=node.split("#")
                    f1.write(nodes[0])
                    f1.write("\t")
                    f1.write(nodes[1])
                    f1.write("\t")
                    f1.write(str(cooc_dict[node]))
                    f1.write("\n")
                    cooc_count=cooc_count+1
        f1.close()
        
        
        #Writing the Co-occurrence file for circos plot input
        writeFile="circos/Voc_Only/"+lineage+"_Cooc.txt"
        with open(writeFile,"w") as f1:
            for node in cooc_dict.keys():
                if cooc_dict[node] >= 0.8:
                    nodes=node.split("#")
                    f1.write("hs1")
                    f1.write(" ")
                    f1.write(str(int(float(nodes[0].split(":")[1]))))
                    f1.write(" ")
                    f1.write(str(int(float(nodes[0].split(":")[1])+1)))
                    f1.write(" ")
                    f1.write("hs1")
                    f1.write(" ")
                    f1.write(str(int(float(nodes[1].split(":")[1]))))
                    f1.write(" ")
                    f1.write(str(int(float(nodes[1].split(":")[1])+1)))
                    f1.write(" ")
                    f1.write("color=green_a2")
                    f1.write(",")
                    f1.write("thickness=")
                    xx=2.57*cooc_dict[node]
                    if xx < 1:
                        xx=1.15
                    f1.write(str(xx))
                    f1.write("p")
                    f1.write("\n")
                    cooc_count=cooc_count+1
        f1.close()




        #Writing the mutation names corresponding to the co-occurrence inputs for circos plot
        writeFile="circos/voc_Only/"+lineage+"_Mutation.txt"
        pos_dict={}
        with open(writeFile,"w") as f1:
            for node in cooc_dict.keys():
                if cooc_dict[node] >= 0.8:
                    nodes=node.split("#")
                    if nodes[0] not in pos_dict.keys():
                        f1.write("hs1")
                        f1.write("\t")
                        f1.write(str(int(float(nodes[0].split(":")[1]))))
                        f1.write("\t")
                        f1.write(str(int(float(nodes[0].split(":")[1])+1)))
                        f1.write("\t")
                        f1.write(nodes[0])
                        f1.write("\t")
                        f1.write("color=chr3")
                        f1.write("\n")
                        pos_dict[nodes[0]]=0
                    if nodes[1] not in pos_dict.keys():
                        f1.write("hs1")
                        f1.write("\t")
                        f1.write(str(int(float(nodes[1].split(":")[1]))))
                        f1.write("\t")
                        f1.write(str(int(float(nodes[1].split(":")[1])+1)))
                        f1.write("\t")
                        f1.write(nodes[1])
                        f1.write("\t")
                        f1.write("color=chr3")
                        f1.write("\n")
                        pos_dict[nodes[1]]=0
                    cooc_count=cooc_count+1
        f1.close()                
```

```python
#Compute the proportion of lineages in a particular mutation present in the co-occurring relation or mutual exclusion relation
sub_dict={}

#the input file combine.txt is a file containing both the co-occurrence and the mutual exclusion relations from the circos plot input file. Each line describes a 
#a mutation label corresponding to a co-occurrence or mutual exclusion relation. The columns are tab sparated (e.g. hs1	240	241	CT:240.0	color=chr3).
with open("circos/combined.txt","r") as f1:
    for line in f1:
        st=line.split("\n")[0].split("\t")
        if line =="":
            continue
        sub_dict[st[3]]=0

lineage_dict={}
#Comet algorithm Input file
#This code will extract the lineages in which a particular mutation label is present
with open("allLineageInput.txt","r") as f1:
    for line in f1:
        st=line.split("\t")
        s=[]
        for i in range(1,len(st)-1):
            s.append(st[i])
        lineage_dict[st[0]]=s
sub_count={}
for key in sub_dict.keys():
    count=0.0
    flag=0
    for l in lineage_dict.keys():
        s=lineage_dict[l]
        if key in s:
            flag=1
            count=count+1.0
    
    if flag ==0:
        print(key)
    
    x=(count/1199)*100
    stt="{0:.2f}".format(x)
    sub_count[key]=stt

#writing the lineage proportion file for circos plot
with open("circos/combined.txt","r") as f2:
    with open("circos/proportionnonvoc.txt","w") as f3:
        for line in f2:
            st = line.split("\n")[0].split("\t")
            mutation=st[3]
            val=sub_count[mutation]
            f3.write(st[0])
            f3.write("\t")
            f3.write(st[1])
            f3.write("\t")
            f3.write(st[2])
            f3.write("\t")
            f3.write(str(val))
            f3.write("\t")
            f3.write("color=chr3")
            f3.write("\n")

f3.close()
        
```
