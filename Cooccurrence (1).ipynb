```python
import os
directory="/Users/prochetasen/graphs/comet_lineage_graph_output/modified_graph/"
for filename in os.listdir(directory):
        lineage=filename.split("_")[0]
        lineageFile="/Users/prochetasen/CometInputData/Nucliotide_Substituitions/CometInput_"+lineage+"_NucleotideSub.txt"
        cooc_dict={}
        maxVal=0
        count=0
        with open(lineageFile,"r") as f:
            for line in f:
                st=line.split("\n")[0].split("\t")
                for i in range(1,len(st)-2):
                    for j in range(i+1,len(st)-1):
                        key1=st[i]+"#"+st[j]
                        key2=st[j]+"#"+st[i]
                        if key1 in cooc_dict.keys():
                            cooc_dict[key1]=cooc_dict[key1]+1
                            if maxVal < cooc_dict[key1]:
                                maxVal=cooc_dict[key1]
                        elif key2 in cooc_dict.keys():
                            cooc_dict[key2]=cooc_dict[key2]+1
                            if maxVal < cooc_dict[key2]:
                                maxVal=cooc_dict[key2]
                        else:
                            cooc_dict[key1]=1
                            if maxVal < cooc_dict[key1]:
                                maxVal=cooc_dict[key1]
                count=count+1


        cooc_count=0
        print(lineage)
        writeFile="/Users/prochetasen/Cooc_Graph/lineage/"+lineage+".txt"
        with open(writeFile,"w") as f1:
            for node in cooc_dict.keys():
                cooc_dict[node]=cooc_dict[node]/count
                if cooc_dict[node] > 0.9:
                    nodes=node.split("#")
                    f1.write(nodes[0])
                    f1.write("\t")
                    f1.write(nodes[1])
                    f1.write("\t")
                    f1.write(str(cooc_dict[node]))
                    f1.write("\n")
                    cooc_count=cooc_count+1
        f1.close()
        
print(cooc_count)
    
                
```
