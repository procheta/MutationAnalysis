```python
mutation_dict={}
lineage_dict={}
seq_dict={}
l_dict={}
count=0


flag=0
with open("/users/prochetasen/updated_mutation_mapper_nucleotide.csv","r") as f:
    for line in f:
        if flag == 0:
            flag=1
            continue
        
        x=[]
        st=line.split("\n")[0].split(",")
        seq=st[4].split("|")[0]
        try:
            lineage=st[4].split("|")[4]
            seq=seq+"_"+st[4].split("|")[4]+"#"+st[4].split("|")[7]
        except:
            lineage=""
            seq=seq+"_NA"+"#"+st[4].split("|")[7]
        mutation=st[0]+st[1]+":"+st[2]
        l_dict[seq]=lineage
        if "-" in mutation:
            continue;
        if seq not in mutation_dict.keys():
            l=[]
            l.append(mutation)
            mutation_dict[seq]=l
        else:
            l=mutation_dict[seq]
            l.append(mutation)
            mutation_dict[seq]=l
        if count%400000==0:
            print("Lines finshed ",count)
        count=count+1
        

count=0
count1=0
print("Total number of Sequencs ", len(mutation_dict.keys()))
with open("/users/prochetasen/LatestCometInput.txt","w") as f:
    for x in mutation_dict.keys():
            l=mutation_dict[x]
            xx=l_dict[x]
            if xx in lineage_dict.keys():
                if lineage_dict[xx] == l:
                    count=count+1
                    continue;
                else:
                    count1=count1+1
            else:
                lineage_dict[xx]=l
            f.write(x)
            f.write("\t")
            for x1 in l:
                f.write(x1)
                seq_dict[x1]=0
                f.write("\t")
            f.write("\n")
f.close()
print("Number of unique lineages ",len(lineage_dict.keys()))  
print("Number of unique substituitions", len(seq_dict.keys()))
            
```
